BFS

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
def bfs (graph,start):
    startNode = start
    visited = list()
    queue = list()

    visited.append(startNode)
    queue.append(startNode)

    while (queue):
        node = queue.pop(0)
        print (f"Node: {node}")
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
```

```python
bfs(graph, "A")
```

DFS

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
def dfs (graph,start):
    startNode = start
    visited = list()
    stack = list() # LIFO


    stack.append(start)

    while (stack):
        node = stack.pop()
        if node is not visited:
            visited.append(node)
            print (f"Node: {node}")
        for neighbor in reversed(graph.get(node,[])):
            if neighbor not in visited:
                stack.append(neighbor)
```

```python
dfs (graph, "A")
```
